**applying multiple filter criteria to a pandas DataFrame**

```python
import pandas as pd
```

```python
movies=pd.read_csv('http://bit.ly/imdbratings')
```

```python
movies.head()
```

```python
movies[(movies.duration>200)&(movies.genre=='Drama')]
```

```python
movies[(movies.duration>200)|(movies.genre=='Adventure')].head()
```

```python
(movies.duration>200)|(movies.genre=='Adventure')
```

```python
movies[movies.genre.isin(['Adventure','Crime'])].head()
```
